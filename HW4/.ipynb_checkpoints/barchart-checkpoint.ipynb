```python
import csv
import matplotlib.pyplot as plt

complaint_types = {}
with open('output_data.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if row[0] not in complaint_types:
            complaint_types[row[0]] = int(row[2])
        else:
            complaint_types[row[0]] += int(row[2])

c = 0
for i,j in complaint_types.items():
    if j>c:
        c = j
        typec = i

print(typec)
x = [typec]
y = [complaint_types[typec]]

plt.bar(x, y, color='blue')
plt.show()


    
```
